```python
import whois
from urllib.parse import urlparse

def is_domain_registered(domain):
    try:
        whois_info = whois.whois(domain)
        return bool(whois_info)
    except Exception as e:
        print(f"Error occurred while checking WHOIS for {domain}: {e}")
        return False

def extract_features(url):
    # Extract domain from URL
    domain = urlparse(url).netloc
    
    # Initialize features dictionary
    features = {
        'UsingIP': 1 if "@" in url else -1,
        'LongURL': 1 if len(url) > 54 else -1,
        'ShortURL': 1 if "bit.ly" in url or "tinyurl.com" in url else -1,
        'Symbol@': 1 if "@" in url else -1,
        'Redirecting//': 1 if "//" in url else -1,
        'PrefixSuffix-': 1 if '-' in url else -1,
        'SubDomains': url.count('.') - 2,
        'HTTPS': 1 if url.startswith("https") else -1,
        'DomainReg': 1 if is_domain_registered(domain) else -1,  # Check if domain is registered
        'Favicon': 1 if "favicon.ico" in url else -1,
        'NonStdPort': 1 if ":" in urlparse(url).netloc else -1,
        'HTTPSDomainURL': 1 if "https://" in urlparse(url).netloc else -1,
        'RequestURL': 1 if "?" in url else -1,
        'AnchorURL': 1 if "http" in url.split("#")[0] else -1,
        'LinksInScriptTags': 1 if "<script>" in url else -1,
        'ServerFormHandler': 1 if "form" in url else -1,
        'InfoEmail': 1 if "mailto:" in url else -1,
        'AbnormalURL': 1 if "https://www." in url or "http://www." in url else -1,
        'WebsiteForwarding': 1 if "forward" in url or "redirect" in url else -1
    }
    return features

# Example user URL
#user_url = "https://www.example.com"

# Extract features from the user URL
features = extract_features(user_url)

# Print the values for the features
for feature, value in features.items():
    print(f"{feature}: {value}")
```

```python
import pickle
import whoose
import gradio as gr

def make_prediction_and_get_output(url):
    # Assuming you have features dictionary extracted from URL
    features = extract_features(url)
    
    # Predict if it's a phishing website
    feature_values = [[value for value in features.values()]] 
    with open("filename.pkl", "rb") as f:
        clf = pickle.load(f)
        preds = clf.predict(feature_values)
    prediction_result = "It is a Phishing Website" if preds == -1 else "It is not a Phishing Website"
    
    # Retrieve domain ownership information
    try:
        d = whoose.who_info(url)
        additional_information = "\n".join([f"Args: {keys},\t\t\t\t Values:{values}" for keys, values in zip(d.keys(), d.values())])
    except whoose.Exception as e:
        additional_information = f"Error retrieving whois information: {e}"
    
    return prediction_result, additional_information


url_input = gr.Textbox(label="Enter URL Name:")
url_input1 = gr.Textbox(label="Rise the report:")
# Define the output components
output1 = gr.Textbox(label="Prediction:")
output2 = gr.Textbox(label="Additional Information:")

# Create the Gradio interface
app = gr.Interface(
    fn=make_prediction_and_get_output, 
    inputs=url_input, 
    outputs=[output1, output2], 
    title="Phishing Website Detector"
)

app.launch()
```
